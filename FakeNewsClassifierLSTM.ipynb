```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
### Vocabulary size
voc_size=5000
```

```python
messages=X.copy()
```

```python
messages.head(1)
```

```python
messages['title'][1]
```

```python
messages
```

```python
messages.reset_index(inplace=True)
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
# https://colab.research.google.com/drive/1CoNxaecjcsAKwDEwNpBmvsKjVFpNTdv0?usp=sharing#scrollTo=WOYLZSrdkQbB
```

```python
nltk.download('stopwords')
```

```python
## Dataset Preprocessing
from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
corpus[1]
```

```python
onehot_repr=[one_hot(words,voc_size) for words in  corpus]
onehot_repr
```

```python
corpus[1]
```

```python
onehot_repr[1]
```

```python
sent_length=20
embaded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embaded_docs)
```

```python
embaded_docs
```

```python
embaded_docs[1]
```

```python
onehot_repr[1]
```

```python
embaded_docs[0]
```

```python
#creating model
model=Sequential() ##name of the model
embedding_vector_feature=40 ##feature presentation[........40 feature for each word]
model.add(Embedding(voc_size,embedding_vector_feature,input_length=sent_length)) ##creting embedding layer
model.add(LSTM(100)) ### LSTM neural network
model.add(Dense(1,activation='sigmoid'))  ## due to output layer is binary e need to add sigmid function
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) #feature of model 
print(model.summary()) ## model summary
```

```python
len(embaded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embaded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
X_final,y_final
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test, y_train, y_test=train_test_split(X_final,y_final,test_size=0.33, random_state=0)
```

```python
##finally training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10, batch_size=64)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred>0.5,1,0) ##AUC RDC Curve
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
